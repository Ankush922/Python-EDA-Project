```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Dell\Downloads\Python_Diwali_Sales_Analysis\Python_Diwali_Sales_Analysis\Diwali Sales Data.csv',  encoding= 'unicode_escape')
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
```

```python
import seaborn as sns
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

* Most of the buyers are females than males. Purchasing power of females is more than males.

## Age

```python
df.columns
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
age = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in age.containers:
    age.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

*From the above graph we can see that most of the buyers from 26-35 age group are females.

## State

```python
#Total number of orders from Top 5 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)

sns.set(rc = {'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'Orders', y = 'State', orient='h')
```

```python
#Total amount of sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data= sales_state, x = 'Amount', y = 'State', orient = 'h')
```

From the above graph we can say that most of the orders are from Utter Pradesh, Maharashtra and Karnataka


## Marital Status

```python
ms = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc = {'figure.figsize':(10,5)})

for bars in ms.containers:
    ms.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

Most of the buyers are married(women) and they have high purchasing power.

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
occ = sns.countplot(data = df, x = 'Occupation')

for bars in occ.containers:
    occ.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False).sum().sort_values(by = 'Amount', ascending=False)

sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')
```

From the above graph we can say that most of the buyers are working in IT, Healthcare and Aviation sector.

## Product Category

```python
sns.set(rc={'figure.figsize': (10,5)})
Prod = sns.countplot(data = df, y = 'Product_Category', orient = 'h')

for bars in Prod.containers:
    Prod.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head()

sns.set(rc={'figure.figsize': (10,5)})
sns.barplot(data = sales_state, x = 'Amount', y = 'Product_Category',orient = 'h')
```

From the above graph we can say that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders')
```

## Conclusion

Mostly the buyers are married women of 28-35 years of age group from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation industry
